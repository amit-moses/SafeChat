```python
# ╔══════════════════════════════════════════════════════════════════════╗
# ║  Toxicity Classifier — Fine-tuning for Cyberbullying Detection      ║
# ║  Dataset : google/civil_comments                                     ║
# ║  Model   : DistilBERT → binary classifier (safe=0 / toxic=1)        ║
# ║  Save    : .pkl (Python-only)  +  ONNX (for React frontend)         ║
# ╚══════════════════════════════════════════════════════════════════════╝
#
# ┌─────────────────────────────────────────────────────────────────────┐
# │  FLAG: LOAD_FROM_MEMORY                                             │
# │  True  → skip training entirely, load model from .pkl, run eval    │
# │  False → train from scratch (overwrites any existing .pkl)         │
# └─────────────────────────────────────────────────────────────────────┘
#
# Instructions (Google Colab):
#   1. Runtime → Change runtime type → T4 GPU
#   2. Paste this entire file into one cell, or split by # ── CELL N ──
#   3. Set LOAD_FROM_MEMORY = True/False in CELL 3
#   4. Run all


# ── CELL 1 ── Install dependencies ────────────────────────────────────────
import subprocess, sys

def install_packages() -> None:
    packages = [
        "transformers", "datasets", "accelerate",
        "scikit-learn", "onnx", "onnxruntime", "optimum[onnxruntime]",
    ]
    subprocess.run([sys.executable, "-m", "pip", "install", "-q"] + packages, check=True)
    print("✅ Packages installed")

install_packages()


# ── CELL 2 ── Imports & GPU check ─────────────────────────────────────────
import os, shutil, json, sys, pickle
import numpy as np
import matplotlib.pyplot as plt
import torch

sys.modules.pop("torchvision", None)
sys.modules.pop("torchvision.io", None)

from datasets import load_dataset
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TrainingArguments,
    Trainer,
    DataCollatorWithPadding,
    TrainerCallback,
    pipeline,
)
from sklearn.metrics import (
    accuracy_score, f1_score, precision_score, recall_score,
)

def check_gpu() -> None:
    if torch.cuda.is_available():
        name   = torch.cuda.get_device_name(0)
        memory = torch.cuda.get_device_properties(0).total_memory / 1e9
        print(f"✅ GPU available: {name}  ({memory:.1f} GB)")
    else:
        print("⚠️  No GPU detected — go to: Runtime → Change runtime type → T4 GPU")

check_gpu()


# ── CELL 3 ── Configuration ───────────────────────────────────────────────

# ┌─────────────────────────────────────────────────────────────────┐
# │  ★  SET THIS FLAG  ★                                           │
# │  True  → load from .pkl, skip training, run synthetic eval      │
# │  False → train from scratch (overwrites existing .pkl)          │
# └─────────────────────────────────────────────────────────────────┘
LOAD_FROM_MEMORY = False   # ← change me

MODEL_NAME    = "distilbert-base-uncased"
MAX_LENGTH    = 128
THRESHOLD     = 0.5
BATCH_SIZE    = 32
EPOCHS        = 6
LEARNING_RATE = 2e-5
SEED          = 42

TRAIN_SAMPLES = 180000
VAL_SAMPLES   =   10000
TEST_SAMPLES  =   40000

# Output paths
OUTPUT_DIR   = "/content/toxicity_model"    # HuggingFace weights (temp)
PKL_PATH     = "/content/toxicity_clf.pkl"  # ← single Python-only save file
ONNX_DIR     = "/content/toxicity_onnx"     # ONNX (for React)
PLOTS_DIR    = "/content/plots"
os.makedirs(PLOTS_DIR, exist_ok=True)

TOXIC_COLS = [
    "toxicity", "severe_toxicity", "obscene",
    "threat", "insult", "identity_attack", "sexually_explicit",
]

print("✅ Configuration loaded")
print(f"   LOAD_FROM_MEMORY = {LOAD_FROM_MEMORY}")
print(f"   PKL file         = {PKL_PATH}")


# ── CELL 4 ── Helpers: save / load .pkl ───────────────────────────────────

def save_clf_to_pkl(clf_pipeline, path: str) -> None:
    """
    Pickle the HuggingFace pipeline so it can be reloaded later without
    re-training.  The .pkl is Python-only — it cannot be used from JS/React
    (use the ONNX export for that).
    """
    with open(path, "wb") as f:
        pickle.dump(clf_pipeline, f)
    size_mb = os.path.getsize(path) / 1e6
    print(f"✅ Model saved → {path}  ({size_mb:.1f} MB)")


def load_clf_from_pkl(path: str):
    """Load a previously pickled classifier pipeline."""
    if not os.path.exists(path):
        raise FileNotFoundError(
            f"No .pkl found at {path}.\n"
            "Set LOAD_FROM_MEMORY = False to train and save the model first."
        )
    with open(path, "rb") as f:
        clf = pickle.load(f)
    print(f"✅ Model loaded from {path}")
    return clf


# ── CELL 5 ── Classifier helpers (needed in both paths) ───────────────────

def load_classifier(model_dir: str):
    """Load the fine-tuned model as a HuggingFace text-classification pipeline."""
    return pipeline(
        "text-classification",
        model=model_dir,
        tokenizer=model_dir,
        truncation=True,
        max_length=MAX_LENGTH,
        device=0 if torch.cuda.is_available() else -1,
    )


def is_toxic(text: str, clf) -> dict:
    result = clf(text)[0]
    return {
        "text":  text,
        "toxic": result["label"] == "toxic",
        "label": result["label"],
        "score": round(result["score"], 4),
    }


# ── CELL 6 ── Synthetic evaluation set ────────────────────────────────────

SYNTHETIC_TOXIC = [
    "I hate you so much, you're disgusting.",
    "Go kill yourself, nobody would miss you.",
    "You are the dumbest person I have ever met.",
    "I will find you and make you regret everything.",
    "You're worthless and you know it.",
    "Shut up you stupid idiot.",
    "You should be ashamed of existing.",
    "Nobody likes you, everyone thinks you're a loser.",
    "You're so ugly it hurts to look at you.",
    "I hope you get hit by a car.",
    "You're a complete waste of space.",
    "Get out of here before I hurt you.",
    "You are absolutely pathetic.",
    "People like you don't deserve to live.",
    "You're a fat disgusting pig.",
    "I'll destroy your life if you don't back off.",
    "You're the worst human being on this planet.",
    "Go crawl back into whatever hole you came from.",
    "You're a brain-dead moron.",
    "Everyone hates you and wishes you were gone.",
    "You make me sick just looking at you.",
    "I will ruin you completely.",
    "You're such a pathetic little crybaby.",
    "Drop dead already.",
    "You're too stupid to understand anything.",
    "I hope something terrible happens to you.",
    "You're garbage and you'll always be garbage.",
    "Nobody will ever love someone as disgusting as you.",
    "You're a joke and everyone is laughing at you.",
    "I want you to suffer.",
    "You are a racist piece of trash.",
    "Idiots like you shouldn't be allowed online.",
    "You deserve every bad thing that happens to you.",
    "Stop breathing, you're wasting oxygen.",
    "You're a coward and everyone can see it.",
    "I'll make your life a living hell.",
    "You're mentally ill and dangerous.",
    "Get a life you pathetic loser.",
    "You're an embarrassment to your family.",
    "I despise everything about you.",
    "You're scum and always will be.",
    "I hope you choke.",
    "Your existence is a mistake.",
    "You're a vile creature.",
    "You should be locked up.",
    "Nobody wants you here, just leave.",
    "You're too dumb to breathe.",
    "I wish you would just disappear forever.",
    "You are absolutely revolting.",
    "Crawl back under your rock, troll.",
]

SYNTHETIC_SAFE = [
    "Have a wonderful day, I hope everything goes well for you!",
    "Can you help me with my homework assignment please?",
    "I really enjoyed reading your article, very insightful.",
    "Thanks so much for your help, I really appreciate it.",
    "The weather today is absolutely beautiful.",
    "I'm feeling a bit sad today but I know things will get better.",
    "Great job on the presentation, you did really well!",
    "I disagree with your opinion but I respect your perspective.",
    "Could you please pass me the salt?",
    "I love spending time with my family on weekends.",
    "This movie was really good, I recommend it.",
    "Let me know if you need any help with the project.",
    "I think we can find a solution if we work together.",
    "Your artwork is absolutely stunning.",
    "I'm proud of how far you've come.",
    "The meeting is scheduled for 3pm tomorrow.",
    "I'd love to learn more about this topic.",
    "Thank you for being so patient with me.",
    "I made a mistake and I'm sorry about that.",
    "This is a really interesting discussion.",
    "I'm not sure I agree but I see your point.",
    "Could we schedule a call to discuss this further?",
    "I hope you feel better soon.",
    "Your feedback was really helpful, thank you.",
    "I just finished a great book, I'd recommend it.",
    "Let's try to solve this problem together.",
    "I really enjoyed our conversation today.",
    "Happy birthday! I hope you have a fantastic day.",
    "The food at that restaurant was delicious.",
    "I think we should consider all options carefully.",
    "I'm so excited about this new opportunity.",
    "Thank you for taking the time to explain that.",
    "I'm going to take a walk in the park later.",
    "This is a complex issue that deserves careful thought.",
    "I appreciate your honesty, even when it's hard to hear.",
    "Let me know if you have any questions.",
    "I'm working on improving myself every day.",
    "Your support means the world to me.",
    "I think we made the right decision.",
    "I'm looking forward to seeing you soon.",
    "It was a tough day but I got through it.",
    "I love how creative this community is.",
    "Can we talk about this calmly?",
    "I think everyone deserves to be treated with respect.",
    "This problem is hard but not impossible.",
    "I'm grateful for everything I have.",
    "Let's agree to disagree on this one.",
    "I just wanted to say you're doing a great job.",
    "I'm here if you need someone to talk to.",
    "What a beautiful sunset tonight.",
]

SYNTHETIC_AMBIGUOUS = [
    "This homework is literally killing me.",
    "I'm dead, that joke was so funny.",
    "Bro I'm gonna destroy you at FIFA tonight.",
    "That presentation absolutely murdered it.",
    "Your idea is total trash, no offense.",
    "I killed it on that exam!",
    "This traffic is making me want to scream.",
    "I could kill for a pizza right now.",
    "That movie was so bad it hurt my brain.",
    "You're dead to me if you spoil the ending.",
    "I want to destroy that buffet tonight.",
    "She absolutely slaughtered that performance.",
    "This song is fire, it's killing everything.",
    "I'm going to murder this workout.",
    "That idea is garbage, come on.",
    "Bro you wrecked him in that argument.",
    "I'm dying over here, this is too funny.",
    "That comeback was brutal, he destroyed you.",
    "This weather is absolutely brutal.",
    "I hate Mondays so much I could scream.",
    "You guys are killing me with these puns.",
    "That plot twist murdered my expectations.",
    "I'm going to annihilate this pizza.",
    "Nobody asked for your opinion on this one.",
    "That fashion choice is a crime against humanity.",
]


def evaluate_synthetic(clf) -> None:
    examples = (
        [(msg, 1, "TOXIC") for msg in SYNTHETIC_TOXIC]  +
        [(msg, 0, "SAFE")  for msg in SYNTHETIC_SAFE]   +
        [(msg, 0, "AMBIG") for msg in SYNTHETIC_AMBIGUOUS]
    )

    all_preds, all_labels, rows = [], [], []
    for msg, true_int, true_str in examples:
        out      = is_toxic(msg, clf)
        pred_int = 1 if out["toxic"] else 0
        pred_str = "TOXIC" if out["toxic"] else "SAFE"
        correct  = pred_int == true_int
        rows.append((msg, true_str, pred_str, out["score"], correct))
        all_preds.append(pred_int)
        all_labels.append(true_int)

    sep  = "-" * 90
    sep2 = "=" * 90
    print("\n" + sep2)
    print(f"  {'#':<4}  {'TRUE':<6}  {'PRED':<6}  {'CONF':>6}  {'OK?':<4}  TEXT")
    print(sep2)
    for i, (msg, true_str, pred_str, score, correct) in enumerate(rows, 1):
        ok   = "OK" if correct else "XX"
        conf = f"{score*100:.1f}%"
        text = msg[:52] + "..." if len(msg) > 52 else msg
        if i == 51:
            print(sep); print("  -- SAFE examples (50) --"); print(sep)
        if i == 101:
            print(sep); print("  -- AMBIGUOUS slang — true label is SAFE (25) --"); print(sep)
        print(f"  {i:<4}  {true_str:<6}  {pred_str:<6}  {conf:>6}  {ok:<4}  {text}")

    acc    = accuracy_score(all_labels, all_preds)
    f1     = f1_score(all_labels, all_preds, zero_division=0)
    prec   = precision_score(all_labels, all_preds, zero_division=0)
    rec    = recall_score(all_labels, all_preds, zero_division=0)
    errors = [(msg, true_str, pred_str, score)
              for msg, true_str, pred_str, score, ok in rows if not ok]
    n_ambig_err = sum(1 for msg, true_str, pred_str, score, ok in rows
                      if not ok and true_str == "AMBIG")

    print(sep2)
    print(f"  RESULTS ({len(rows)} examples) | Accuracy: {acc*100:.1f}% | F1: {f1*100:.1f}% | "
          f"Precision: {prec*100:.1f}% | Recall: {rec*100:.1f}% | Errors: {len(errors)}/{len(rows)}")
    print(f"  Ambiguous slang over-triggered: {n_ambig_err}/{len(SYNTHETIC_AMBIGUOUS)}")
    print(sep2)
    if errors:
        print("\n  Misclassified:")
        for msg, true_str, pred_str, score in errors:
            print(f"    True={true_str}  Pred={pred_str}  Conf={score*100:.1f}%  \"{msg[:65]}\"")
    print("")


# ══════════════════════════════════════════════════════════════════════════
#  MAIN BRANCH — controlled by LOAD_FROM_MEMORY flag
# ══════════════════════════════════════════════════════════════════════════

if LOAD_FROM_MEMORY:
    # ── PATH A: load existing model, skip training ─────────────────────
    print("=" * 60)
    print("  LOAD_FROM_MEMORY = True")
    print("  Loading model from .pkl …")
    print("=" * 60)

    clf = load_clf_from_pkl(PKL_PATH)
    print("\n🧪 Running synthetic evaluation on loaded model …")
    evaluate_synthetic(clf)

else:
    # ── PATH B: train from scratch ─────────────────────────────────────
    print("=" * 60)
    print("  LOAD_FROM_MEMORY = False → training from scratch")
    print("=" * 60)

    # ── CELL B1 ── Data ────────────────────────────────────────────────

    def make_binary_label(example):
        scores = [example.get(col, 0.0) or 0.0 for col in TOXIC_COLS]
        example["label"] = int(any(s > THRESHOLD for s in scores))
        return example

    # Computed split boundaries from the constants — no magic numbers anywhere
    SPLIT_END_TRAIN = TRAIN_SAMPLES
    SPLIT_END_VAL   = TRAIN_SAMPLES + VAL_SAMPLES
    SPLIT_END_TEST  = TRAIN_SAMPLES + VAL_SAMPLES + TEST_SAMPLES

    def load_and_prepare():
        """
        Concatenate all three civil_comments splits, shuffle once, then
        slice into train / val / test using the constants defined in CELL 3.
        This guarantees zero overlap between splits regardless of how the
        original dataset was partitioned.
        """
        from datasets import concatenate_datasets

        print("📦 Loading civil_comments …")
        ds = load_dataset("google/civil_comments")
        total_available = sum(len(ds[s]) for s in ("train", "validation", "test"))
        print(f"   Available — Train: {len(ds['train']):,} | "
              f"Val: {len(ds['validation']):,} | Test: {len(ds['test']):,} | "
              f"Total: {total_available:,}")

        needed = SPLIT_END_TEST
        if needed > total_available:
            raise ValueError(
                f"Requested {needed:,} examples "
                f"(TRAIN {TRAIN_SAMPLES:,} + VAL {VAL_SAMPLES:,} + TEST {TEST_SAMPLES:,}) "
                f"but dataset only has {total_available:,}."
            )

        print(f"\n   Merging all splits and shuffling (seed={SEED}) …")
        full = concatenate_datasets(
            [ds["train"], ds["validation"], ds["test"]]
        ).shuffle(seed=SEED)

        print(f"   Slicing → train[0:{SPLIT_END_TRAIN:,}]  "
              f"val[{SPLIT_END_TRAIN:,}:{SPLIT_END_VAL:,}]  "
              f"test[{SPLIT_END_VAL:,}:{SPLIT_END_TEST:,}]")

        train = full.select(range(0,            SPLIT_END_TRAIN)).map(make_binary_label)
        val   = full.select(range(SPLIT_END_TRAIN, SPLIT_END_VAL)).map(make_binary_label)
        test  = full.select(range(SPLIT_END_VAL,   SPLIT_END_TEST)).map(make_binary_label)

        # Sanity-check: zero overlap
        train_texts = set(train["text"])
        val_texts   = set(val["text"])
        test_texts  = set(test["text"])
        tv = len(train_texts & val_texts)
        tt = len(train_texts & test_texts)
        vt = len(val_texts   & test_texts)
        print(f"\n   ✅ Overlap check — train∩val: {tv}  train∩test: {tt}  val∩test: {vt}")

        toxic = sum(train["label"])
        total = len(train)
        print(f"   Class balance (train) — "
              f"Toxic: {toxic:,} ({100*toxic/total:.1f}%)  "
              f"Safe: {total-toxic:,} ({100*(total-toxic)/total:.1f}%)")
        return train, val, test

    train_raw, val_raw, test_raw = load_and_prepare()

    # ── CELL B2 ── Tokenise ────────────────────────────────────────────

    print(f"🔤 Loading tokenizer: {MODEL_NAME}")
    tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)

    def build_tokenized_splits(train_raw, val_raw, test_raw):
        def tok(batch):
            return tokenizer(batch["text"], truncation=True,
                             max_length=MAX_LENGTH, padding=False)
        keep = ["input_ids", "attention_mask", "label"]
        print("⚙️  Tokenizing …")
        tr = train_raw.map(tok, batched=True).select_columns(keep)
        va = val_raw.map(tok,   batched=True).select_columns(keep)
        te = test_raw.map(tok,  batched=True).select_columns(keep)
        for ds in (tr, va, te):
            ds.set_format("torch", columns=keep)
        print("✅ Done")
        return tr, va, te

    train_ds, val_ds, test_ds = build_tokenized_splits(train_raw, val_raw, test_raw)

    # ── CELL B3 ── Model ───────────────────────────────────────────────

    model = AutoModelForSequenceClassification.from_pretrained(
        MODEL_NAME, num_labels=2,
        id2label={0: "safe", 1: "toxic"},
        label2id={"safe": 0, "toxic": 1},
    )
    total_params = sum(p.numel() for p in model.parameters())
    print(f"✅ Model: {MODEL_NAME}  ({total_params/1e6:.1f} M params)")

    # ── CELL B4 ── Training ────────────────────────────────────────────

    training_history = {
        "epoch": [], "train_loss": [], "val_loss": [],
        "accuracy": [], "f1": [], "precision": [], "recall": [],
        "accuracy_steps": [],
    }

    class HistoryCallback(TrainerCallback):
        PRINT_EVERY_N_STEPS = 50

        def __init__(self):
            self._train_start = self._epoch_start = self._total_steps = None

        def on_train_begin(self, args, state, control, **kwargs):
            import time
            self._train_start = time.time()
            self._total_steps = state.max_steps
            print(f"\n{'═'*60}\n  Training started | Steps: {self._total_steps:,} | "
                  f"Epochs: {int(args.num_train_epochs)}\n{'═'*60}\n")

        def on_epoch_begin(self, args, state, control, **kwargs):
            import time
            self._epoch_start = time.time()
            print(f"\n{'─'*60}\n  Epoch {int(state.epoch)+1}/{int(args.num_train_epochs)} started\n{'─'*60}")

        def on_log(self, args, state, control, logs=None, **kwargs):
            import time
            if not logs: return
            step    = state.global_step
            total   = self._total_steps or 1
            elapsed = time.time() - (self._train_start or time.time())
            eta     = (elapsed / step * (total - step) / 60) if step > 0 else 0
            if "loss" in logs:
                loss = logs["loss"]
                training_history["train_loss"].append({"step": step, "loss": loss})
                if step % self.PRINT_EVERY_N_STEPS == 0 or step == 1:
                    print(f"  Step {step:>5}/{total} [{100*step/total:5.1f}%]  "
                          f"loss={loss:.4f}  elapsed={elapsed/60:.1f}m  ETA≈{eta:.1f}m")

        def on_epoch_end(self, args, state, control, **kwargs):
            import time
            t = time.time() - (self._epoch_start or time.time())
            print(f"\n  Epoch {int(state.epoch)} done in {t/60:.1f} min — validating …")

        def on_evaluate(self, args, state, control, metrics=None, **kwargs):
            if not metrics: return
            epoch = int(state.epoch)
            training_history["epoch"].append(epoch)
            training_history["val_loss"].append(metrics.get("eval_loss"))
            training_history["accuracy"].append(metrics.get("eval_accuracy"))
            training_history["f1"].append(metrics.get("eval_f1"))
            training_history["precision"].append(metrics.get("eval_precision"))
            training_history["recall"].append(metrics.get("eval_recall"))
            training_history["accuracy_steps"].append(
                {"step": state.global_step, "acc": metrics.get("eval_accuracy", 0)})
            acc  = metrics.get("eval_accuracy", 0)
            f1   = metrics.get("eval_f1", 0)
            prec = metrics.get("eval_precision", 0)
            rec  = metrics.get("eval_recall", 0)
            loss = metrics.get("eval_loss", "—")
            print(f"\n  ┌─ Epoch {epoch} Validation {'─'*30}┐")
            print(f"  │  Loss={loss:.4f}  Acc={acc*100:.2f}%  F1={f1*100:.2f}%  "
                  f"Prec={prec*100:.2f}%  Rec={rec*100:.2f}%")
            print(f"  └{'─'*48}┘\n")

        def on_train_end(self, args, state, control, **kwargs):
            import time
            mins = (time.time() - (self._train_start or time.time())) / 60
            print(f"\n{'═'*60}\n  Training finished in {mins:.1f} min\n{'═'*60}\n")

    def compute_metrics(eval_pred):
        logits, labels = eval_pred
        preds = np.argmax(logits, axis=-1)
        return {
            "accuracy":  accuracy_score(labels, preds),
            "f1":        f1_score(labels, preds, zero_division=0),
            "precision": precision_score(labels, preds, zero_division=0),
            "recall":    recall_score(labels, preds, zero_division=0),
        }

    def compute_class_weights(train_ds):
        train_ds.reset_format()
        labels  = [int(x) for x in train_ds["label"]]
        train_ds.set_format("torch", columns=["input_ids", "attention_mask", "label"])
        total   = len(labels)
        n_safe  = labels.count(0)
        n_toxic = labels.count(1)
        w_safe  = total / (2 * n_safe)
        w_toxic = total / (2 * n_toxic)
        print(f"   ⚖️  Weights — safe: {w_safe:.3f}  toxic: {w_toxic:.3f}  "
              f"(toxic penalty ×{w_toxic/w_safe:.0f})")
        device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
        return torch.tensor([w_safe, w_toxic], dtype=torch.float).to(device)

    class WeightedTrainer(Trainer):
        def __init__(self, *args, class_weights=None, **kwargs):
            super().__init__(*args, **kwargs)
            self.class_weights = class_weights

        def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
            labels  = inputs.pop("labels")
            outputs = model(**inputs)
            loss    = torch.nn.CrossEntropyLoss(weight=self.class_weights)(
                          outputs.logits, labels)
            return (loss, outputs) if return_outputs else loss

    import inspect
    tok_kwarg = ("processing_class"
                 if "processing_class" in inspect.signature(Trainer.__init__).parameters
                 else "tokenizer")

    args = TrainingArguments(
        output_dir=OUTPUT_DIR,
        num_train_epochs=EPOCHS,
        per_device_train_batch_size=BATCH_SIZE,
        per_device_eval_batch_size=BATCH_SIZE,
        learning_rate=LEARNING_RATE,
        weight_decay=0.01,
        warmup_ratio=0.1,
        eval_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model="f1",
        logging_steps=50,
        fp16=torch.cuda.is_available(),
        seed=SEED,
        report_to="none",
    )

    print("⚖️  Computing class weights …")
    weights = compute_class_weights(train_ds)

    trainer = WeightedTrainer(
        model=model,
        args=args,
        train_dataset=train_ds,
        eval_dataset=val_ds,
        **{tok_kwarg: tokenizer},
        data_collator=DataCollatorWithPadding(tokenizer),
        compute_metrics=compute_metrics,
        callbacks=[HistoryCallback()],
        class_weights=weights,
    )

    print("🚀 Starting training … (~60 min on T4 GPU)")
    trainer.train()
    print("✅ Training complete!")

    # ── CELL B5 ── Save HF model + pickle ────────────────────────────

    trainer.save_model(OUTPUT_DIR)
    tokenizer.save_pretrained(OUTPUT_DIR)
    print(f"✅ HuggingFace model saved → {OUTPUT_DIR}")

    clf = load_classifier(OUTPUT_DIR)       # build pipeline from saved weights
    save_clf_to_pkl(clf, PKL_PATH)          # pickle it → single .pkl file

    # ── CELL B6 ── Plots ──────────────────────────────────────────────

    def plot_loss_per_step(history, save_dir):
        entries = history["train_loss"]
        if not entries:
            print("No loss data."); return
        steps  = [e["step"] for e in entries]
        losses = [e["loss"]  for e in entries]
        fig, ax = plt.subplots(figsize=(12, 4))
        ax.plot(steps, losses, alpha=0.3, color="#E74C3C", linewidth=0.8, label="Raw")
        w = max(1, len(losses) // 60)
        smooth = np.convolve(losses, np.ones(w)/w, mode="valid")
        ax.plot(steps[:len(smooth)], smooth, color="#C0392B", linewidth=2,
                label=f"Smoothed (w={w})")
        ax.set_title("Training Loss per Step", fontsize=14, fontweight="bold")
        ax.set_xlabel("Global Step"); ax.set_ylabel("Loss"); ax.legend()
        plt.tight_layout()
        path = os.path.join(save_dir, "loss_per_step.png")
        plt.savefig(path, dpi=150, bbox_inches="tight"); plt.show()
        print(f"📉 Loss plot → {path}")

    def plot_accuracy_per_step(history, save_dir):
        acc_entries = history.get("accuracy_steps", [])
        if not acc_entries:
            print("No accuracy data."); return
        steps = [e["step"] for e in acc_entries]
        accs  = [e["acc"]  for e in acc_entries]
        fig, ax = plt.subplots(figsize=(12, 4))
        ax.plot(steps, accs, "o-", color="#9B59B6", linewidth=2, markersize=7)
        for s, a in zip(steps, accs):
            ax.annotate(f"{a*100:.1f}%", xy=(s, a), xytext=(0, 10),
                        textcoords="offset points", ha="center", fontsize=10)
        ax.set_title("Validation Accuracy per Step", fontsize=14, fontweight="bold")
        ax.set_xlabel("Global Step"); ax.set_ylabel("Accuracy"); ax.set_ylim(0, 1)
        plt.tight_layout()
        path = os.path.join(save_dir, "accuracy_per_step.png")
        plt.savefig(path, dpi=150, bbox_inches="tight"); plt.show()
        print(f"📈 Accuracy plot → {path}")

    plot_loss_per_step(training_history, PLOTS_DIR)
    plot_accuracy_per_step(training_history, PLOTS_DIR)

    # ── CELL B7 ── Export ONNX (for React) ───────────────────────────

    def export_to_onnx(hf_model_dir, onnx_dir):
        from optimum.onnxruntime import ORTModelForSequenceClassification
        print("🔄 Exporting ONNX …")
        os.makedirs(onnx_dir, exist_ok=True)
        ort = ORTModelForSequenceClassification.from_pretrained(
            hf_model_dir, export=True)
        ort.save_pretrained(onnx_dir)
        tokenizer.save_pretrained(onnx_dir)
        meta = {
            "model_type": "distilbert", "task": "text-classification",
            "id2label": {"0": "safe", "1": "toxic"},
            "label2id": {"safe": 0, "toxic": 1},
            "max_length": MAX_LENGTH, "threshold": THRESHOLD,
        }
        with open(os.path.join(onnx_dir, "react_meta.json"), "w") as f:
            json.dump(meta, f, indent=2)
        print(f"✅ ONNX → {onnx_dir}  Files: {os.listdir(onnx_dir)}")

    export_to_onnx(OUTPUT_DIR, ONNX_DIR)

    # ── CELL B8 ── Synthetic evaluation ───────────────────────────────
    print("\n🧪 Running synthetic evaluation …")
    evaluate_synthetic(clf)

    # ── CELL B9 ── Download everything ────────────────────────────────

    def download_all():
        from google.colab import files
        # Zip the ONNX dir (for React)
        shutil.make_archive("/content/toxicity_onnx", "zip", ONNX_DIR)
        files.download("/content/toxicity_onnx.zip")
        # Zip the plots
        shutil.make_archive("/content/toxicity_plots", "zip", PLOTS_DIR)
        files.download("/content/toxicity_plots.zip")
        # Download the single pkl (Python model)
        files.download(PKL_PATH)
        print("📦 Downloads triggered: onnx.zip, plots.zip, toxicity_clf.pkl")

    download_all()


# ── CELL 7 ── Quick demo (always runs) ────────────────────────────────────

def run_demo(clf) -> None:
    examples = [
        "I hate you, you're so stupid and ugly!",
        "Have a wonderful day, hope you feel better!",
        "you should go kill yourself",
        "Can you help me with my homework please?",
        "Nobody likes you, you're a complete loser",
        "Great job on the presentation today!",
        "I will find you and make you regret this",
        "Thanks so much for your help, really appreciate it!",
    ]
    print("\n🧪 Quick demo:")
    print("=" * 65)
    for msg in examples:
        out  = is_toxic(msg, clf)
        icon = "🚫 TOXIC" if out["toxic"] else "✅ SAFE "
        print(f"{icon}  [{out['score']:.2f}]  {msg[:55]}")

run_demo(clf)


# ── CELL 8 ── Test a custom message ───────────────────────────────────────

def check_custom(clf, message: str) -> None:
    out = is_toxic(message, clf)
    print(f"Message : {out['text']}")
    print(f"Result  : {'🚫 Toxic' if out['toxic'] else '✅ Safe'}")
    print(f"Score   : {out['score'] * 100:.1f}%")

check_custom(clf, "Write your message here")   # ← change me
```
